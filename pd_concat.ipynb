```python
import pandas as pd
```

```python
# warningという警告文が表示されないようになる
import warnings
warnings.filterwarnings('ignore')
```

```python
df01 = pd.DataFrame( {'氏名':['佐藤','鈴木','高橋','田中'],
                      'クラス':['df01','df01','df01','df01'],
                      '数学':[1, 2, 3, 4],
                      '国語':[5, 6, 7, 8]})
df01
```

```python
df02 = pd.DataFrame( {'氏名':['佐藤','鈴木','山本'],
                      'クラス':['df02','df02','df02'],
                      '数学':[9, 10, 11],
                      '国語':[12, 13, 14]})
df02
```

```python
# concatで縦方向に結合
pd.concat([df01, df02])
```

```python
# index_numberを振りなおす
pd.concat([df01, df02], ignore_index=True)
```

```python
df03 = pd.DataFrame( {'氏名':['佐藤','鈴木','加藤'],
                      'クラス':['df03','df03','df03'],
                      '数学':[15, 16, 17],
                      '国語':[18, 19, 20]})
df03
```

```python
# 3つ以上でも同時につなげることができる
pd.concat([df01, df02, df03])
```

```python
# マルチインデックスを作成
df = pd.concat([df01, df02, df03],keys=['1番目','2番目','3番目'])
df
```

```python
df.index
```

```python
df04 = pd.DataFrame( {'氏名':['佐藤','鈴木','佐々木'],
                      'クラス':['df04','df04','df04'],
                      '数学':[21, 22, 23],
                      '社会':[24, 25, 26]})
df04
```

```python
# 共通項以外のコラムがあっても繋ぐことができる
pd.concat([df01, df04])
```

```python
# defaultはouter
pd.concat([df01, df04], join='outer')
```

```python
# 共通カラムのみ出力
pd.concat([df01, df04], join='inner')
```

```python
# axis=0で縦(default)
pd.concat([df01, df04], axis=0)
```

```python
# axis='index'でも縦
pd.concat([df01, df04], axis='index')
```

```python
# concatでもaxis=1で横を指定すればmergeのように横方向にも結合可能
pd.concat([df01, df04], axis=1)
```

```python
# 1の代わりにcolumnsでもok(デフォルトではindexがkey)
pd.concat([df01, df04],axis='columns')
```

```python
pd.concat([df01,df02,df03,df04], axis='columns')
```

```python
pd.concat([df01, df02, df03, df04],axis='columns',join='inner')
```

```python
# indexを氏名に変更
df01 = df01.set_index('氏名')
df02 = df02.set_index('氏名')
df03 = df03.set_index('氏名')
df04 = df04.set_index('氏名')
```

```python
df04
```

```python
# 共通するindexのみ結合し出力
pd.concat([df01, df02, df03, df04], axis='columns', join='inner')
```
