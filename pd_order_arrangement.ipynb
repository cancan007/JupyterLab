```python
#df = pd.read_excel(r'C:\Users\User\Documents\Data\data.xlsx', "sheet_name=''")  #シート名があれば、そっちでも指定できる
df = pd.read_excel(r'C:\Users\User\Documents\Data\data.xls')        #.xlsx形式だと、IndexError: list index out of rangeが起こるため、 .xls に変更
df
```

```python
df.dtypes
```

```python
del df['注']
```

```python
df['人口'] = pd.to_numeric(df['人口（総数）'], errors='coerce')
df['人口（男性数）'] = pd.to_numeric(df['人口（男）'], errors='coerce')
df['人口（女性数）'] = pd.to_numeric(df['人口（女）'], errors='coerce')
df
```

```python
df.dtypes
```

```python
df = df.drop(['人口（総数）', '人口（女）', '人口（男）'], axis=1)   #複数のcolumnsを消去
```

```python
df
```

```python
df = df.dropna(how='any')
```

```python
df
```

```python
df.sort_values(by='人口')     #指定したコラムで、昇順にする。（昇順：小さいものから大きいものへ, あ～ん, a~z）
```

```python
df.sort_values(by='人口', ascending=False)        #　降順
```

```python
df.sort_values(by=['都道府県名', '人口'], ascending='True')  #都道府県名ごとに人口が昇順　（昇順ならascending='True'は書く必要ない）
```

```python
df.sort_values(by=['都道府県名','人口'], ascending=[False,True])  #都道府県名:降順, 人口:昇順
```

```python
df.sort_values(by='都道府県名',ascending=False,na_position='first')  #指定したコラムに欠損値があれば、それを先頭にする（今回はない）
```

```python
df_ax = pd.DataFrame({'col01':[1, 5, 9], 'col02':[2, 4, 8], 'col03':[3, 6, 7]}
                    ,index=['idx01', 'idx02', 'idx03'])
df_ax
```

```python
df_ax.sort_values(by='idx01', ascending=False, axis=1)      #axis=1 は行単位という意味
```

```python
df_ax.sort_values(by='col01',ascending=False,axis=0)        #axis=0　は列単位という意味
```

```python
df_ax.sort_values(by='idx02',ascending=False,axis=1)
```
