```python
#バランシング(後半の0を消していく方法)
# do balancing
target_count = 0
index_for_balancing = []
for i in range(try_data['target'].shape[0]):
    if try_data['target'][i] == 0:    #0の時のみ反応
        target_count += 1
        if target_count > try_data['target'].sum():
            index_for_balancing.append(i)
            
try_data = try_data.drop(index=index_for_balancing, axis=0)
try_data = try_data.reset_index(drop=True)      # 後のバランシングをするために、インデックスを振りなおす
try_data
```

```python
# データをシャッフル
# shuffle data
shuffled_indices = np.arange(scaled_inputs.shape[0])
np.random.shuffle(shuffled_indices)

shuffled_inputs = scaled_inputs.reindex(index=shuffled_indices)
shuffled_targets = targets_data.reindex(index=shuffled_indices)

#F値を相対的に評価して、不要なコラムを削除する
shuffled_inputs = shuffled_inputs.drop(['cat3','cat5','cat8','cat10','cont0','cont4','cont7','cont9','cont10'],axis=1)
shuffled_targets = shuffled_targets.drop(['cat3','cat5','cat8','cat10','cont0','cont4','cont7','cont9','cont10'],axis=1)
```

```python
# 欠損値がないことを確認(本当は初めにやらなければならない)
# make sure whether null exist or not
shuffled_inputs.isnull().any(axis=0)
```

```python
p_count = shuffled_inputs.shape[0]

train_p_count = int(0.8 * p_count)
validation_p_count = p_count - train_p_count

train_p_inputs = shuffled_inputs[:train_p_count]
train_p_targets = shuffled_targets[:train_p_count]

validation_p_inputs = shuffled_inputs[train_p_count:]
validation_p_targets = shuffled_targets[train_p_count:]
```

```python
import tensorflow as tf
```

```python
# 作ったnpzデータをロードして代入
# loading npz files
npz = np.load('Tabular_train.npz')
train_inputs = npz['inputs'].astype(np.float)
train_targets = npz['targets'].astype(np.int)

npz = np.load('Tabular_validation.npz')
validation_inputs = npz['inputs'].astype(np.float)
validation_targets = npz['targets'].astype(np.int)

npz = np.load('Tabular_test.npz')
test_inputs = npz['inputs'].astype(np.float)
```

```python
# モデルの設定を行う
# create model
inputs_size=21
outputs_size=2
hidden_layer_size=50

model = tf.keras.Sequential([
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(outputs_size, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

batch_size=100
max_epochs=100

early_stopping = tf.keras.callbacks.EarlyStopping(patience=2)

model.fit(train_inputs,
          train_targets,
          batch_size=batch_size,
          epochs=max_epochs,
          callbacks=[early_stopping],
          validation_data = (validation_inputs, validation_targets),
          verbose=2
         )
```

```python
model.predict(test_inputs)  #左が0である確率、右が1である確率
```

```python
model.evaluate(validation_inputs, validation_targets)
```
