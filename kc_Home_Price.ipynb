```python
def label_trans(self):
    for column in self.select_dtypes(include=object).columns:
        le = preprocessing.LabelEncoder()
        le.fit(self[column])
        self[column] = le.transform(self[column])
```

```python
scaler = StandardScaler()
def point_scale(self):
    for column in un_label_columns:
        scaler.fit(self[column].values.reshape(-1,1))
        self[column] = scaler.transform(self[column].values.reshape(-1,1))
```

```python
def search_nan_and_fill(self):
    nan_columns = []
    #def __init__(self):
        #self.nan_columns = nan_columns
    for column in self.columns:
        if self[column].isnull().any():    #nanを含むコラムにTrueを返す
            nan_columns.append(column)
            
    for n_column in nan_columns:
        self[n_column] = self[n_column].fillna(self[n_column].median())
```

```python
# REGRESSION MODELS
lm = LinearRegression().fit(xSTrain, yTrain)
pls = PLSRegression().fit(xSTrain, yTrain)     # https://punhundon-lifeshift.com/pca_pls :サンプルサイズが小さかったり、多重共線性が考えられる場合に有効な分析方法の一つ
ridge =  Ridge().fit(xSTrain, yTrain)   #https://qiita.com/K_Noguchi/items/3f5cf527d6f6d46767fb#:~:text=%E3%83%AA%E3%83%83%E3%82%B8%E5%9B%9E%E5%B8%B0%E3%81%A8%E3%81%AF%E9%81%8E,%E4%B8%80%E3%81%A4%E3%81%A8%E3%82%82%E8%A8%80%E3%81%88%E3%81%BE%E3%81%99%E3%80%82 過学習を抑える手法の一つ 
lasso = Lasso().fit(xSTrain, yTrain)   # いらない特徴量を削る
elasticnet = ElasticNet().fit(xSTrain, yTrain)    # https://leck-tech.com/machine-learning/elastic-net#Lasso Ridge回帰とLasso回帰のハイブリッドのような形を取っているElasticNet
knnr = KNeighborsRegressor().fit(xSTrain, yTrain)   # k最近傍法
cartr = DecisionTreeRegressor(random_state=42).fit(xSTrain, yTrain)   # 決定木の回帰木 (説明URLはrandomforestにあり)
baggr = BaggingRegressor(random_state=42, bootstrap_features=True, verbose=False).fit(xSTrain, yTrain)   # バギング法(説明URLはGradientBoostingにあり)
rfr = RandomForestRegressor(random_state=42, verbose=False).fit(xSTrain, yTrain)     # https://qiita.com/yshi12/items/6d30010b353b084b3749   アンサンブル学習法(複数の分類器を集めて構成される分類器)の一つ
gbmr = GradientBoostingRegressor(verbose=False).fit(xSTrain, yTrain)       # https://qiita.com/nazoking@github/items/51a46256ecda598b60dd  勾配ツリーブースト または勾配ブースト回帰ツリー（GBRT）は、任意の微分可能な損失関数にブーストする一般化
xgbr = XGBRegressor().fit(xSTrain, yTrain)    # http://tekenuko.hatenablog.com/entry/2016/09/22/220814
lgbmr = LGBMRegressor().fit(xSTrain, yTrain)    # https://www.codexa.net/lightgbm-beginner/  
catbr = CatBoostRegressor(verbose=False).fit(xSTrain, yTrain)   # https://ryucoding.com/programming/catboost-beginner  カテゴリー変数に強いXGBoostやLightGBMに並ぶ勾配ブースディングの一つ
```

```python
models = [lm,pls,ridge,lasso,elasticnet,knnr,cartr,baggr,rfr,gbmr,xgbr,lgbmr,catbr]
```
