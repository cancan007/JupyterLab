```python
import pandas as pd
```

```python
df01 = pd.DataFrame( {'氏名':['高橋', '伊藤', '鈴木', '佐藤'],
                      'クラス':['a001', 'a001', 'a001', 'a001'],
                      '数学':[1, 2, 3, 4]})
df01
```

```python
df02 = pd.DataFrame( {'氏名':['高橋', '伊藤', '鈴木', '佐藤'],
                      'クラス':['a001', 'a001', 'a001', 'a001'],
                      '国語':[5, 6, 7, 8]})
df02
```

```python
# mergeでonで指定したものをもとに、結合
pd.merge(df01, df02, on='氏名')
```

```python
# suffixes でmerge時に　コラム名を指定してつなげる
pd.merge(df01, df02, on='氏名', suffixes=['_left', '_right'])
```

```python
pd.merge(df01, df02, on='氏名', suffixes=['', '_重複'])
```

```python
df02_name = pd.DataFrame( {'名前':['高橋', '伊藤', '鈴木', '佐藤'],
                      'クラス':['a001', 'a001', 'a001', 'a001'],
                      '国語':[5, 6, 7, 8]})
df02_name
```

```python
# コラム名が違っていても、下のように指定して結合可能
pd.merge(df01, df02_name, left_on='氏名', right_on='名前')
```

```python
df03 = pd.DataFrame( {'名前':['高橋', '高橋'],
                      'クラス':['a001', 'a002'],
                      '英語':[15, 16]})
df03
```

```python
# 氏名とクラス、名前とクラスで、それぞれ同じものと結合
pd.merge(df01, df03, left_on=['氏名','クラス'], right_on=['名前','クラス'])
```

```python
df04 = pd.DataFrame( {'氏名':['高橋','伊藤','渡辺','加藤'],
                      'クラス':['a001','a001','a001','a001'],
                      '国語':[5, 6, 7, 8]})
df04
```

```python
# 左のデータがベース(結合時、左はすべて残る)
pd.merge(df01, df04, on='氏名', how='left')
```

```python
# 右がベース
pd.merge(df01, df04, on='氏名', how='right')
```

```python
# outerで共通項以外も結合時に消えずに出力
pd.merge(df01, df04, on='氏名', how='outer')
```

```python
# 元のデータフレームの情報を開示
pd.merge(df01, df04, on='氏名', how='outer', indicator=True)
```

```python
# inner で共通項のみ出力
pd.merge(df01, df04, on='氏名', how='inner')
```

```python
df05 = pd.DataFrame({'クラス':['a001','a001','a001','a001'],
                     '数学':[1, 2, 3, 4]},
                    index=['高橋','伊藤','鈴木','佐藤'])
df05
```

```python
df02
```

```python
# 左のデータフレームのインデックスと右のデータフレームの氏名をキーに結合
pd.merge(df05, df02, left_index=True, right_on='氏名')
```

```python
# 結合したデータフレームを結合キーでソートして表示(df05(index) と df02(氏名) の結合キーの昇順で並べる)
pd.merge(df05, df02, left_index=True, right_on='氏名', sort=True)
```
